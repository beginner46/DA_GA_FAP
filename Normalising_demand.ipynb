```python
df_8_12 = pd.read_csv('8_12.csv')
df_4_8 = pd.read_csv('4_8.csv')
df_12_6 = pd.read_csv('12_16.csv')
```

Here time travel has been normalized between [0,1] to go well with the genetic algorithm another reason for doing this was that the density lies in the range of [0,1] and for the algorithm to work properly it was required traveltime to have the same distribution as density.

```python
result_8_12 = df_8_12.apply(lambda iterator: ((iterator.max() - iterator)/(iterator.max() - iterator.min())).round(2))
result_4_8 = df_4_8.apply(lambda iterator: ((iterator.max() - iterator)/(iterator.max() - iterator.min())).round(2))
result_12_16 = df_12_6.apply(lambda iterator: ((iterator.max() - iterator)/(iterator.max() - iterator.min())).round(2))
```
